```python
import matplotlib.pyplot as plt

import json
import requests
import pandas as pd

#Gather data for FL

#Palm Beach, FL
#26.7056° N, 80.0364° W

# See available parameters and parameter definition at https://docs.oikolab.com/#3-weather-parameters
params= {'param': ['temperature',  
                   'total_precipitation'],
          'start': '2015-01-01',                
          'end': '2017-12-31',
          'lat': 26.71,        # Data will be interpolated to specified lat/lon
          'lon': -80.04,
          'freq': 'D',         # 'H' for hourly, 'D' for daily data
          'api-key': '68784205c8ff4242a5350e144401f224'
        }
```

```python
#Load data from FL into dataframe

r = requests.get('https://api.oikolab.com/weather', params=params)

weather_data = json.loads(r.json()['data'])
df_FL = pd.DataFrame(index = pd.to_datetime(weather_data['index'],unit = 's'),
                  data = weather_data['data'],
                  columns = weather_data['columns'])
df_FL.index.name = 'datetime'

print('Processing time: %s seconds'% r.json()['attributes']['processing_time'])
```

```python
df_FL.head()
```

```python
#Clean up column names and include only temperature and precipitation

df_FL2 = df_FL[['temperature (degC)', 'total_precipitation (mm of water equivalent)']]

df_FL2.reset_index(inplace=True)

df_FL_clean = df_FL2.rename(columns = {'datetime':'Date','temperature (degC)':'Temperature (C)', 
                                     'total_precipitation (mm of water equivalent)': 'Precipitation (mm)'})
df_FL_clean.head()
```

```python
#Get average temp and precipitation by month

data_month_FL = df_FL_clean.resample('M', on='Date').mean()
data_month_FL.head()
```
