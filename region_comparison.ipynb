```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#FL csv
FL_file = "Resources/FL_weather_data2.csv"
FL_df = pd.read_csv(FL_file)

#TX csv
TX_file = "Resources/TX_weather_data2.csv"
TX_df = pd.read_csv(TX_file)

#LA csv
LA_file = "Resources/LA_weather_data2.csv"
LA_df = pd.read_csv(LA_file)
```

```python
FL2 = pd.to_datetime(FL_df['Date'])
```

```python
FL3 = FL2.dt.strftime('%Y-%m')
```

```python
#All Temps

#FL

x_axis = FL3
FL = FL_df['Temperature (C)']
TX = TX_df['Temperature (C)']
LA = LA_df['Temperature (C)']

plt.plot(x_axis, FL_df['Temperature (C)'])
plt.xlabel('Month in 2015-2017')
plt.ylabel('Temperature (C)')
plt.title('Monthly Temperatures: Palm Beach, FL 2015-2017')

#TX
plt.plot(x_axis, TX_df['Temperature (C)'])
plt.xlabel('Month in 2015-2017')
plt.ylabel('Temperature (C)')
plt.title('Monthly Temperatures: Corpus Christi 2015-2017')

#lA
plt.plot(x_axis, LA_df['Temperature (C)'])
plt.xlabel('Month in 2015-2017')
plt.ylabel('Temperature (C)')
plt.title('Monthly Temperature: Iberville, LA 2015-2017')

tick_locations = [value for value in x_axis]
plt.xticks(tick_locations, x_axis, rotation=45)


FL_handle = plt.plot(x_axis, FL, marker ='o', color='blue', label="Florida")
TX_handle = plt.plot(x_axis, TX, marker='o', color='red', label="Texas")
TX_handle = plt.plot(x_axis, LA, marker='o', color='green', label="Louisana")

plt.legend(loc="lower right")

plt.rcParams['figure.figsize'] = [12,8]

```
