```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#California csv

CA_file = "Resources/CA_weather_data.csv"
```

```python
CA_df = pd.read_csv(CA_file)
CA_df.head()
```

```python
#Set variables for plotting

CA_temp = CA_df['Temperature (C)']
x_axis1 = CA_df['Date']
```

```python
#Plot average temp 

CA_avg1 = plt.plot(x_axis1, CA_temp, color="blue", label="Avg Temperature (C)")

tick_locations = [value for value in x_axis1]
plt.xticks(tick_locations, x_axis1, rotation=90)

#Formatting
plt.title("Average Temp(C): San Diego, CA")
plt.xlabel("Months")
plt.ylabel("Temperature (C)")

plt.rcParams['figure.figsize'] = [8,6]

# Save 
plt.savefig("Images/CA_temp.png")

#Show

plt.show()
```

```python
#Precipitation variables

CA_rain = CA_df['Precipitation (mm)']
x_axis2 = CA_df['Date']
```

```python
#Plot average precipitation

CA_avg2 = plt.plot(x_axis2, CA_rain, color="green", label="Avg Precipitation (mm)")

tick_locations = [value for value in x_axis2]
plt.xticks(tick_locations, x_axis2, rotation=90)

#Formatting
plt.title("Avg Precipitation (mm): San Diego, CA")
plt.xlabel("Months")
plt.ylabel("Precipitation (mm)")

plt.rcParams['figure.figsize'] = [8,6]

# Save 
plt.savefig("Images/CA_rain.png")

#Show
plt.show()
```
