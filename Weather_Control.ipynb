```python
import matplotlib.pyplot as plt

import json
import requests
import pandas as pd

#San Diego, CA
#32.7157° N, 117.1611° W

# See available parameters and parameter definition at https://docs.oikolab.com/#3-weather-parameters
params= {'param': ['temperature',  
                   'total_precipitation'],
          'start': '2015-01-01',                
          'end': '2017-12-31',
          #'lat': 32.71,        # Data will be interpolated to specified lat/lon
          #'lon': -117.16,
          'location': 91945,
          'freq': 'D',         # 'H' for hourly, 'D' for daily data
         #Brian's Api Key 
         'api-key': 'd00f29c170ce449c93d7f89c59288bf5'
        }
```

```python
r = requests.get('https://api.oikolab.com/weather', params=params)

weather_data = json.loads(r.json()['data'])
df_CA = pd.DataFrame(index = pd.to_datetime(weather_data['index'],unit = 's'),
                  data = weather_data['data'],
                  columns = weather_data['columns'])
df_CA.index.name = 'datetime'

print('Processing time: %s seconds'% r.json()['attributes']['processing_time'])
```

```python
df_CA.head()
```

```python
#Clean up column names and include only temperature and precipitation

df_CA2 = df_CA[['temperature (degC)', 'total_precipitation (mm of water equivalent)']]

df_CA2.reset_index(inplace=True)

df_CA_clean = df_CA2.rename(columns = {'datetime':'Date','temperature (degC)':'Temperature (C)', 
                                     'total_precipitation (mm of water equivalent)': 'Precipitation (mm)'})
df_CA_clean.head()
```

```python
#Get average temp and precipitation by month

data_month_CA = df_CA_clean.resample('M', on='Date').mean()
data_month_CA.head()

```

```python
#Export df into CSV file for later use (avoid using API key for multiple GET requests)

data_month_CA.to_csv("CA_weather_data.csv", index=True, header=True)
```
