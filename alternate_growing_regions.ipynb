```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Florida csv

FL_file = "Resources/FL_weather_data2.csv"
```

```python
FL_df = pd.read_csv(FL_file)
FL_df.head()
```

```python
#Reformat date column
```

```python
#Set variables for plotting

FL_temp = FL_df['Temperature (C)']
x_axis1 = FL_df['Date']
```

```python
#Plot average temp 

FL_avg1 = plt.plot(x_axis1, FL_temp, color="blue", label="Avg Temperature (C)")

tick_locations = [value for value in x_axis1]
plt.xticks(tick_locations, x_axis1, rotation=90)

#Formatting
plt.title("Average Temp(C): Palm Beach, FL")
plt.xlabel("Months")
plt.ylabel("Temperature (C)")

plt.rcParams['figure.figsize'] = [8,6]

#Show

plt.show()
```

```python
FL_rain = FL_df['Precipitation (mm)']
x_axis2 = FL_df['Date']
```

```python
#Plot average precipitation

FL_avg2 = plt.plot(x_axis2, FL_rain, color="green", label="Avg Precipitation (mm)")

tick_locations = [value for value in x_axis2]
plt.xticks(tick_locations, x_axis2, rotation=90)

#Formatting
plt.title("Avg Precipitation (mm): Palm Beach, FL")
plt.xlabel("Months")
plt.ylabel("Precipitation (mm)")

plt.rcParams['figure.figsize'] = [8,6]

#Show
plt.show()
```
