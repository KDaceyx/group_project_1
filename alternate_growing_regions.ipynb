```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Florida csv

FL_file = "Resources/FL_weather_data2.csv"
```

```python
FL_df = pd.read_csv(FL_file)
FL_df
```

```python
#Reformat date column
```

```python
#Set variables for plotting

FL_temp = FL_df['Temperature (C)']
x_axis1 = FL_df['Date']
```

```python
#Plot average temp 

plt.plot(FL_df['Date'], FL_df['Temperature (C)'])
plt.xlabel('Month in 2015-2017')
plt.ylabel('Temperature (C)')
plt.title('Monthly Temperatures: Palm Beach, FL 2015-2017')

tick_locations = [value for value in x_axis1]
plt.xticks(tick_locations, x_axis1, rotation=90)

plt.rcParams['figure.figsize'] = [12,8]

# Save 
plt.savefig("Images/FL_temp.png")

#Show

plt.show()
```

```python
#Precipitation variables

FL_rain = FL_df['Precipitation (mm)']
x_axis2 = FL_df['Date']
```

```python
#Plot average precipitation

plt.plot(FL_df['Date'], FL_df['Precipitation (mm)'])
plt.xlabel('Month in 2015-2017')
plt.ylabel('Precipitation (mm)')
plt.title('Monthly Precipitation: Palm Beach, FL 2015-2017')

tick_locations2 = [value for value in x_axis2]
plt.xticks(tick_locations2, x_axis2, rotation=90)

plt.rcParams['figure.figsize'] = [12,8]

# Save 
plt.savefig("Images/FL_rain.png")

#Show
plt.show()
```
